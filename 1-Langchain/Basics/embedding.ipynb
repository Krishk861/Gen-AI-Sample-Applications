```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="This is a tutorial on OpenAI embedding"
query_text=embeddings.embed_query(text)
query_text
```

```python

from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")
loader
text_document=loader.load()
text_document
```

```python
## Storing vector in vector db
!pip install chromadb
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(text_document)
```

```python
query="At the stroke of the midnight hour, when the world sleeps, India will awake to life and freedom. "
retrived_results=db.similarity_search(query)
print(retrived_results)
```
